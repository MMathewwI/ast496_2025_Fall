```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
# last line only needed to make Jupyter play nice with plots
```

```python
x = np.arange(0, 51, 1)
y = 10*np.exp(-0.1*x) + np.sin(0.5*x)
```

```python
print(y)
```

```python
plt.plot(x, y, 'k-')
```

```python
plt.plot(x, y, 'ro', label=r'$y = 10e^{-0.1x} + \sin(\frac{x}{2})$')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot([10, 20, 30, 40, 50, 60]) # show plot and return immediately
plt.show(block=False) 
```

# You can modify plots you've already created

```python
plt.ylabel('new label')      # oops! forgot to add y-label
plt.draw()                   # updates current plot window
line3 = plt.plot([2, 4, 5, 3, 1, 0]) # add a new plot
plt.draw()                   # update plot window
```

# You can even modify elements on plots you've created like our orange line - see whats going to happen to it...

```python
plt.get(line3)
```

```python
plt.setp(line3, linestyle='dashed', lw=4, color='crimson')
```

```python
plt.figure(figsize=(5,3))
x = np.random.randn(1000)
y = np.random.randn(1000)
colors = np.random.rand(1000)
sizes = 50*np.random.randn(1000)
plt.scatter(x, y, s=sizes, c=colors, cmap='turbo', edgecolor='black')
plt.grid()
ax = plt.gca()
ax.set_xlabel('x')
ax.set_ylabel('y')
fig = plt.gcf()
title = fig.suptitle('blah', fontsize=20)
title.set_fontname('Arial')
title.set_fontstyle('oblique')
title.set_color('C1')
```

```python
plt.figure()
ax2 = plt.gcf().add_subplot()
x = np.linspace(0, 2*np.pi, 20)
y = np.sin(x)
y_errors = np.abs(0.1*np.cos(x))
ax2.errorbar(x, y, yerr=y_errors, fmt='.-')
```

```python
plt.figure(figsize=(5, 3))
x = np.linspace(0, 4*np.pi, 1000)
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.plot(x, np.tan(x), label='_nolegend_')
plt.legend(loc='upper left')
plt.ylim(-5, 5)
plt.text(np.pi, -2, r'Random Greek: $\alpha = \int_0^1 f(x) dx$', fontsize=16, rotation=45)
```
